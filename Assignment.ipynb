In my analysis I will analyze into the patterns of sugar-sweetened beverage consumption in California residents on a daily basis. The data I will be analyzing for adolescents was taken in 2012, and for adults and children the data is from the year 2013. I will delve in deeper whether being obese has anything to do with drinking more beverages as well as if more income in adolescents and adults has to do with more drinking. The last thing I wanted to analyze was if education had an effect on consumptions in adults vs if parent education has an effect on how much sugar-sweetened beverages the child was consuming. 

### importing Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading and renaming categories. Please double check that sugar.csv is in the same directory.

```python
# Read the data
sugar = pd.read_csv("sugar.csv")
#renaming the columns
sugar.columns = ['year', 'Age Group', 'Category', 'Type', 'Average servings per day']
#renamin Adolescent (12-17) to Adolescent
sugar['Age Group'] = sugar['Age Group'].replace('Adolescent (12-17)', 'Adolescent')
#renamin Adult (18+) to Adult
sugar['Age Group'] = sugar['Age Group'].replace('Adult (18+)', 'Adult')
#renamin Child (6-11) to Child  
sugar['Age Group'] = sugar['Age Group'].replace('Child (6-11)', 'Child')
```

First thing I wanted to compare was if being people are overweight do they have the tendency to drink more beverages vs if they were less. My assumption would be that if they are overweight they have a tendency to have more servings per day.

```python
overweight = sugar[sugar['Category'] == 'Overweight Status']

p = sns.barplot(x='Age Group', y='Average servings per day', hue='Type', data=overweight)
p = plt.xlabel('Age Group')
p = plt.ylabel('Average servings per day')
p = plt.title('Average servings of sugar-sweetened beverages per day by age group and overweight status')
plt.show()
```

From the data above we know that people that are obese also seem to consume more sugar sweetened beverages. Next lets analyze what effect income has on adolescents.

```python
# print barplot of adolescent and their income vs servings
values = sugar[sugar['Age Group'] == 'Adolescent']
values = values[values['Category'] == 'Income']

p= sns.barplot(x='Type', y='Average servings per day', hue='Category', data=values)
p.tick_params(labelsize=10, axis='x', rotation=45)
p.set_xlabel('Income', fontsize=10)
p.set_title('Average servings of sugar-sweetened beverages per day in adolescents by income')

plt.show()
```

From this graph we can see that the average servings of sugar-sweetened beverages per day in adolescents by income is higher in the low income group than the high income group. This is a bit surprising because we would expect the high income group to have more money to spend on sugar-sweetened beverages. However, this could be because the low income group is more likely to buy cheaper sugar-sweetened beverages, which are more likely to be unhealthy.

```python
values = sugar[sugar['Age Group'] == 'Adult']
values = values[values['Category'] == 'Income']

p= sns.barplot(x='Type', y='Average servings per day', hue='Category', data=values)
p.tick_params(labelsize=10, axis='x', rotation=45)
p.set_xlabel('Income', fontsize=10)
p.set_title('Average servings of sugar-sweetened beverages per day in Adults by income')

plt.show()
```

From this graph of adults, it shows the same pattern as adolescents. The less the income is the more servings they have a day. One more thing I wanted to investigate was the correlation of education level to drinking sugar beverages in adults vs if parent's education level show similar relation to child and their consumption.

```python
Education = sugar[sugar['Category'] == 'Education']

p = sns.barplot(x='Type', y='Average servings per day', hue='Category', data=Education, palette='dark:#A9D6E5')
p.tick_params(labelsize=7, axis='x')
p = plt.xlabel('Age Group')
p = plt.ylabel('Average servings per day')
p = plt.title('Average servings of sugar-sweetened beverages per day by age group and Education')

plt.show()
```

```python
Education = sugar[sugar['Category'] == 'Parent Education']

p = sns.barplot(x='Type', y='Average servings per day', hue='Category', data=Education)
p = plt.xlabel('Age Group')
p = plt.ylabel('Average servings per day')
p = plt.title('Average servings of sugar-sweetened beverages per day by age group and Education')

plt.show()
```

Using the two graphs above we can infer that the further education a person did the less likely they were to drink sugar sweetened beverages.
